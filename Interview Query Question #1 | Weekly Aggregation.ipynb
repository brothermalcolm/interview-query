Given a list of timestamps in sequential order, return a list of lists grouped by week (7 days) using the first timestamp as the starting point.

```python
# Example:

ts = [
    '2019-01-01', 
    '2019-01-02',
    '2019-01-08', 
    '2019-02-01', 
    '2019-02-02',
    '2019-02-05',
]

output = [
    ['2019-01-01', '2019-01-02'], 
    ['2019-01-08'], 
    ['2019-02-01', '2019-02-02'],
    ['2019-02-05'],
]
```

```python
# initialize
from datetime import datetime, timedelta

# find the first timestamp
start = datetime.strptime(ts[0], '%Y-%M-%d')

# create week series
ws = [start + i * timedelta(days=7) for i in range(1,52)]

# create time series
ts.sort()

# for every week, find dates that fall within that week
for w in ws:
    for t in ts:
        week = []
        date = datetime.strptime(t, '%Y-%M-%d')
        if date >= w and date < w + timedelta(weeks=1):
            week.append(date)
        output.append(week)
    if week == []: 
        break

output
```
